```python
import random
import numpy as np
from message import Message
from tqdm import tqdm
from user import User
from numpy.linalg import norm
```

```python
user_0 = User('U0', 'normal user', quality_params=(0.5, 0.15, 0, 1), post_per_day=1, followers=[], friends=[], interests=[1,0,0,0.4,0.6])
```

```python
messages = []

for i in tqdm(range(20)):
    mid = f"P0_U{i+1}"
    uid = f"U{i+1}"
    quality_params = (0.5, 0.15, 0, 1)
    topics = [round(random.uniform(0, 1), 2) for _ in range(5)]
    is_shadow = False
    exposure = []
    
    msg = Message(mid, uid, quality_params, topics, is_shadow, exposure)
    messages.append(msg)
```

```python
for msg in messages:
    print(msg.appeal)
```

```python
user_vector = user_0.interests
similarity_list = []
for message in messages:
    message_vector = message.topics
    similarity = np.dot(user_vector, message_vector) / (norm(user_vector) * norm(message_vector))
    similarity_list.append((message, similarity))
sorted_similarity_list = sorted(similarity_list, key=lambda x: x[1], reverse=True)
sorted_similarity_list[:10]
# Get only the top 10 messages
user_0.feed.extend([i[0] for i in sorted_similarity_list[:10]])
user_0.feed
```
